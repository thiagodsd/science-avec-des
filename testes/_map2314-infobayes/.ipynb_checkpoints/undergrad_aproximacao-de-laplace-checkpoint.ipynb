```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

#print(plt.style.available)

#plt.style.use('ggplot')
#plt.style.use('fivethirtyeight')
#plt.style.use('classic')
plt.style.use('seaborn-deep')
%matplotlib inline
#%matplotlib notebook

import pandas as pd
import numpy as np
import seaborn as sn
```

```python
def P(l, r):
    return np.exp(-l) * (l**r / np.math.factorial(r))

def Q(l, r):
    return np.exp((-(l - r + 1)**2) / (2*(r - 1))) / np.sqrt(2*np.pi*(r - 1))
```

# $\lambda$ pequeno

```python
for r in range(2, 25, 5):
    l = np.linspace(0, 30, 31)
    plt.hold(True)
    plt.stem(l, P(l,r), markerfmt=' ' , c='b', label='Distribution')
    plt.plot(l, Q(l,r), c='r', label="Laplace's Approx.")
    plt.legend(loc='best')
    plt.show()
    plt.clf()
```

# $\lambda$ grande

```python
for r in range(2, 125, 25):
    l = np.linspace(0, 150, 151)
    plt.hold(True)
    plt.stem(l, P(l,r), markerfmt=' ' , c='b', label='Distribution')
    plt.plot(l, Q(l,r), c='r', label="Laplace's Approx.")
    plt.legend(loc='best')
    plt.show()
    plt.clf()
```
