```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

plt.style.use('seaborn-deep')
%matplotlib inline

import pandas as pd
import numpy as np

from keras.models import Sequential
from keras.layers.core import Activation, Dense
```

```python
%matplotlib inline
EPOCH = 2000

data0 = np.array([[0,0],
                  [0,1],
                  [1,1],
                  [1,0]])
target0 = np.array([[1],
                   [0],
                   [1],
                   [0]])

#plt.clf()
#plt.figure(figsize=(10,8))
#plt.scatter(data0[:,0], data0[:,1], c=target0, cmap='gray', s=45)
#plt.show();

# single layered perceptron
model = Sequential()
model.add(Dense(50, input_dim=2, activation='tanh',kernel_initializer='random_uniform',
                bias_initializer='zeros'))
model.add(Dense(30, activation='tanh',kernel_initializer='random_uniform',
                bias_initializer='zeros'))
model.add(Dense(1, activation='sigmoid',kernel_initializer='random_uniform',
                bias_initializer='zeros'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(data0, target0, nb_epoch=EPOCH, verbose=0)
print(model.predict(data0))

#

x_min, x_max = data0[:,0].min()-0.1, data0[:,0].max()+0.1
y_min, y_max = data0[:,1].min()-0.1, data0[:,1].max()+0.1

spacing = min(x_max - x_min, y_max - y_min) / 500
XX, YY = np.meshgrid(np.arange(x_min, x_max, spacing),
                     np.arange(y_min, y_max, spacing))

artificial_data = np.hstack((XX.ravel().reshape(-1,1), 
                             YY.ravel().reshape(-1,1)))

prob = model.predict(artificial_data)
clf = np.where(prob < 0.5, 0, 1)
z = clf.reshape(XX.shape)

plt.clf()
plt.figure(figsize=(10,8))
plt.contourf(XX, YY, z, cmap='gray', alpha=0.5)
plt.scatter(data0[:,0], data0[:,1], c=target0, cmap='gray', s=45)
plt.show()
```
