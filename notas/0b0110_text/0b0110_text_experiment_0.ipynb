# functions

```python
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    # Replace LaTeX-like accents with correct Portuguese characters
    replacements = {
        r'\s?˜a': 'ã',
        r'\s?˜o': 'õ',
        r'\s?´a': 'á',
        r'\s?´e': 'é',
        r'\s?´i': 'í',
        r'\s?´o': 'ó',
        r'\s?´u': 'ú',
        r'\s?`a': 'à',
        r'\s?ˆa': 'â',
        r'\s?ˆe': 'ê',
        r'\s?ˆi': 'î',
        r'\s?ˆo': 'ô',
        r'\s?ˆu': 'û',
        r'\s?ç': 'ç',
        r'\s?´C': 'Ç',
        r'\s?´A': 'Á',
        r'\s?´E': 'É',
        r'\s?´I': 'Í',
        r'\s?´O': 'Ó',
        r'\s?´U': 'Ú',
        r'\s?˜A': 'Ã',
        r'\s?˜O': 'Õ',
        r'\s?ˆA': 'Â',
        r'\s?ˆE': 'Ê',
        r'\s?ˆI': 'Î',
        r'\s?ˆO': 'Ô',
        r'\s?ˆU': 'Û',
    }
    for pattern, replacement in replacements.items():
        text = re.sub(pattern, replacement, text)
    # Handle other specific issues
    text = text.replace("´", "´")
    return text
```
